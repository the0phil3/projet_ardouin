```python
boats = df[df['function']=='Bateau']
boat_list = boats['unittitle'].value_counts()
boat_list[0:30]
```

```python
hero = boats[boats['unittitle']=="Héros, le vaisseau le"]
hero.drop_duplicates(subset='contenu', inplace=True)
```

```python
hero
```

```python
m = folium.Map(location=[46, 2], zoom_start=5)
```

```python
output = pd.DataFrame()
for i, row in hero.iterrows():
    try:
        print(i)
        data = []
        sample = hero['geoJson'].loc[i]
        sample = json.loads(sample)
        loc = sample['features'][0]['geometry']['coordinates']
        loc = loc[::-1]
        name = sample['features'][0]['properties']['name']
        date = row['dateD']
        data = [name, loc, date]
        data = pd.DataFrame(data).transpose()
        output = output.append(data)
    except:
        pass
```

```python
output
```

```python
output = output.set_axis(['name','coordindates', 'dates'],axis=1, inplace=False)
output = output.reset_index(drop= True) 
#output = output.drop_duplicates(subset='name')
```

```python
for i in range(len(output)):
    try:
        folium.CircleMarker(
        location= output.iloc[i][1],
        radius=10,
        popup=output.iloc[i][0],
        color="#3186cc",
        fill=False,
        fill_color="#3186cc",
    ).add_to(m)
    except:
        pass
```

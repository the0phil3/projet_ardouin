```python
import pandas as pd
import numpy as np 

from display_xml import XML
import cython

from perdido.geocoder import Geocoder
geocoder = Geocoder(sources=['geonames'])
```

```python
db = pd.read_csv('/Users/Theo/dev/projet_ardouin/xl2sqlite/output/v.2/controlaccess.tsv', sep = '\t')
db = db[db['geogname'].notna()]
```

```python
def geocode(df):
    geocoder = Geocoder()
    df['geoJson'] = df['geogname'].apply(lambda x: geocoder(x).geojson)
    return df
```

```python
new_db = geocode(db)
```

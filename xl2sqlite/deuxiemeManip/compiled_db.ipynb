```python
total = pd.merge(base, model,on="unittitle")
```

```python
total['function'] = total.apply(lambda row: Type_column(row), axis=1)
```

```python
Personne = total[total['result']==0]
Bateau = total[total['result']==1]
Affaire = total[total['result']==2]
Personne.loc[Personne['function'].str.contains("Personne"),'ID'] = Personne.groupby(['unittitle','function']).ngroup()
Personne.loc[Personne['function'].str.contains("Personne"),'ID'] = "P" + Personne['ID'].astype(str)

Bateau.loc[Bateau['function'].str.contains("Bateau"),'ID'] = Bateau.groupby(['unittitle','function']).ngroup()
Bateau.loc[Bateau['function'].str.contains("Bateau"),'ID'] = "B" + Bateau['ID'].astype(str)

Affaire.loc[Affaire['function'].str.contains("Affaire"),'ID'] = Affaire.groupby(['unittitle','function']).ngroup()
Affaire.loc[Affaire['function'].str.contains("Affaire"),'ID'] = "A" + Affaire['ID'].astype(str)
```

```python
total = pd.concat([Personne,Bateau,Affaire])
```
